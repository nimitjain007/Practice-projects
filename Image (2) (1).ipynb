```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import time
```

```python
driver = webdriver.Chrome(r"C:\Users\Lenovo\Downloads\chromedriver_win32 (1)\chromedriver.exe")
```

```python
url = 'https://www.amazon.com/s?k=jeans+for+men&crid=2S4IN7ZOB3MPT&sprefix=jeans%2Caps%2C377&ref=nb_sb_ss_ts-doa-p_1_5'
driver.get(url)
```

```python
# Collecting url's
urls = []

base_url = "https://www.amazon.in/s?k=Sarees&page={}&qid=1632112444&ref=sr_pg_{}"

for i in range(2,10):
    
    driver.get(base_url.format(i,i))
    
    for url in driver.find_elements_by_xpath("//a[@class='a-link-normal s-no-outline']"):
        
        urls.append(url.get_attribute("href"))
```

```python
# Grabbing image source

images_sarees = []

for link in urls:
    
    driver.get(link)
    time.sleep(1)
    
    try:
        
        img = driver.find_element_by_css_selector("#landingImage")
    
        source= img.get_attribute('src')
    
        images_sarees.append(source)
        
    except NoSuchElementException:
        
        images_sarees.append("Image Not Found!")
```
