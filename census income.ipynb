```python
# converting sex and income to "0" and "1" 
def Income_to_numeric(x):
    if x=='>50K':
        return 1
    if x=='<=50K':
        return 0
    
cdata['Income'] = cdata['Income'].apply(Income_to_numeric)
def Sex_to_numeric(x):
    if x=='Male':
        return 1
    if x=='Female':
        return 0
    
cdata['Sex'] = cdata['Sex'].apply(Sex_to_numeric)
```

```python
X = dumydata.drop(['Income'],axis=1)    
Y = dumydata['Income']
```

```python
X = (X-X.min())/(X.max()-X.min())
X.head()
```

```python
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold(0.005)
X = selector.fit_transform(X)
X.shape,Y.shape
```

#Pca

```python
from sklearn.preprocessing import StandardScaler
```

```python
x_Std=StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
```

```python
principalComponents = pca.fit_transform(x_Std)
```

```python
principalDf = pd.DataFrame(data = principalComponents)
```

```python
principalDf
```

```python
from sklearn.model_selection import train_test_split
principalDf_train,principalDf_test,Y_train,Y_test = train_test_split(principalDf,Y,test_size = 0.2)
```
